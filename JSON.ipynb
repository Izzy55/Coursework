```python
# Question 2 - Find the top 10 major project themes (using column 'mjtheme_namecode')

#Isolated the column
wbdict = worldbank['mjtheme_namecode']
wbdict
```

```python
#Create empty list and use for loop to create list of all names
allthemes = []
for name in wbdict:
    for i in range(len(name)):
        allthemes.append(name[i]['name'])
        
allthemes
```

```python
#Turn list into series
themeseries = pd.Series(allthemes)
```

```python
#find the number of times each name appears in the series - answer to question 2
themeseries.value_counts()
```

```python
#Question 3 - Fill in the empty names

#Created a dictionary of key,value pairs that pairs codes with names
wb = worldbank['mjtheme_namecode']

codesdict = {}
for code in wb:
    for i in range(len(code)):
        if code[i]['code'] not in codesdict.keys():
            if code[i]['name'] != '':
                codesdict[code[i]['code']] = code[i]['name']

codesdict
```

```python
#Identify each empty name, match the code with the corresponding dictionary key from codesdict, assign the value name 
for name in worldbank['mjtheme_namecode']:
    for i in range(len(name)):
        if name[i]['name'] == '':
            name[i]['name'] = codesdict[name[i]['code']]
            
```
